Qual o número de clusters mais adequado para os dados em agrupamento_Q1.csv, e quais os
centroides desses clusters, usando os centroides em agrup_centroides_Q1.csv e KMeans?

Passo 1 - Importação das bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from sklearn.cluster import KMeans
```

Passo 2 - Criação e Carregamento do dataframe

```python
agrupamento_df=pd.read_csv('/content/agrupamento_Q1.csv')
```

```python
agrupamento_df
```

```python
agrupamento_df.head()
```

```python
agrupamento_df.info()
```

```python
agrupamento_df.describe()
```

Passo 3 - Criação e verificação da quantidade necessária de clusters

```python
wcss_1 = []
range_values = range(1, 40)
for i in range_values:
  kmeans = KMeans(n_clusters=i)
  kmeans.fit(agrupamento_df)
  wcss_1.append(kmeans.inertia_)
```

```python
print(wcss_1)
```

```python
plt.plot(wcss_1,'bx-')
plt.xlabel('Clusters')
plt.ylabel('WCSS');
```

Passo 4 - Uso do KMeans para determinação dos clusters

```python
kmeans = KMeans(n_clusters=6)
kmeans.fit(agrupamento_df)
labels = kmeans.labels_
```

```python
labels, len(labels)
```

```python
np.unique(labels, return_counts=True)
```

Passo 5 - Determinação dos centróides

```python
kmeans.cluster_centers_
```

```python
cluster_centers = pd.DataFrame(data = kmeans.cluster_centers_, columns = [agrupamento_df.columns])
cluster_centers
```

```python
agrupamento_df_cluster = pd.concat([agrupamento_df, pd.DataFrame({'cluster': labels})], axis = 1)
agrupamento_df_cluster.head()
```

Passo 6 - Gráficos dos clusters

```python
for i in agrupamento_df.columns:
  plt.figure(figsize=(35,8))
  for j in range(6):
    plt.subplot(1, 6, j + 1)
    cluster = agrupamento_df_cluster[agrupamento_df_cluster['cluster'] == j]
    cluster[i].hist(bins = 20)
    plt.title('{} \nCluster {}'.format(i, j))
  plt.show()
```

Passo 7 - Criação do arquivo ordenado com os clusters

```python
agrupamento_ordered = agrupamento_df_cluster.sort_values(by = 'cluster')
agrupamento_ordered.head()
```

```python
agrupamento_ordered.to_csv('cluster.csv')
```
